```python
N = 50
steps = 1500000
# J = 1
beta = 1
```

```python
Jv = np.random.normal(size=(N,N))
Jh = np.random.normal(size=(N,N))
```

```python
H = np.zeros(steps)
M = np.zeros(steps)
# sigma = np.zeros((steps, N, N))

s = np.random.choice([-1,1], (N,N))

# sigma[0] = s
M[0] = np.sum(s)
for i in range(N):
    for  j in range(N):
        H[0] += -1 * s[i][j] * (Jh[i][j]*s[i][(j+1) % N] + Jv[i][j]*s[(i+1) % 50][j])

# s = np.ones((N, N))
# H[0] = -2 * J * N ** 2
# M[0] = N ** 2
```

```python
fig = plt.figure(figsize=(5, 5))

ax = plt.axes()
ax.set_axis_off()

ax.imshow(s, interpolation='none',cmap='PuBuGn')
```

```python
for i in tqdm(range(1, steps)):
    x, y = np.random.randint(0, N-1, 2)
    dH = 2 * s[x,y] * (s[(x+1) % N, y % N]*Jv[x,y] + s[(x-1+N) % N, y % N]*Jv[(x-1 + N)%N,y] + s[x % N, (y-1+N) % N]*Jh[x % N, (y-1+N) % N] + s[x % N, (y+1) % N]*Jh[x,y]) 
    if np.random.random(1) <= min(np.exp(-beta * dH), 1):
        s[x, y] = -s[x, y]
        H[i] = H[i-1] + dH
        M[i] = M[i-1] + 2 * s[x, y]
    else:
        H[i] = H[i-1]
        M[i] = M[i-1]
    # sigma[i] = s
```

```python
def autocorr_helper(times, maxtime, diff):
    s = 0
    
    flag = 1
    # if diff > times[-1]: flag = -flag
        
    timesSort = np.hstack([times[times>=diff], (times + diff)[(times + diff) < maxtime]])
    timesSort = np.append(timesSort, maxtime)
    timesSort = np.sort(timesSort)
    
    i = 0
    
    while i < len(times) and times[i] < diff:
        flag = -flag
        i += 1
        
        
    prev = diff
        
    for time in timesSort:
        s += (time - prev) * flag
        flag = -flag
        prev = time    
        
    #print(timesSort, end=' ')
    
    return s
```

```python
a = [np.array([1, 2]), 3, 1]
autocorr_helper(*a)
```

```python
for d in range(5):
    print(d, autocorr_helper(np.array([0, 1, 2, 3, 4, 5]), 5, d))
```
